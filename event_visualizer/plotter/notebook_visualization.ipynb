```python
el = EventLoader('/data_cms_upgrade/cuisset/testbeam18/clue3d/v31/cmssw/data/CLUE_clusters.root')
```

```python
event = el.loadEvent(EventID(250, 650, 7))
```

```python
vis_3d = (Clue3DVisualization(event, title=None, projection="perspective", zAspectRatio=1, useLayerAsZ=True)
    .add3DClusters()
    .add2DClusters(True)
    .addRechits(chainAsIndividualGroup=True)
    .addImpactTrajectory()
    .addDetectorCylinder()
    .addSliders()
)
vis_3d.fig.update_layout(dict(scene_camera_projection_type="orthographic"))
vis_3d.fig.show("browser")
```
