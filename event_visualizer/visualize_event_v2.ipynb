```python
evt_array = tree_array[0]

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.view_init(vertical_axis="y")
ax.scatter(evt_array["clus2D_x"],evt_array["clus2D_y"],evt_array["clus2D_z"], s=evt_array["clus2D_energy"]*10,alpha=0.4,c='r',label="2D cluster")
ax.scatter(evt_array["clus3D_x"],evt_array["clus3D_y"],evt_array["clus3D_z"], s=evt_array["clus3D_energy"]*10,alpha=0.4,c='g',label="3D cluster")
ax.set_xlabel("X (cm)")
ax.set_ylabel("Y (cm)")
ax.set_zlabel("Z (cm)")
```
